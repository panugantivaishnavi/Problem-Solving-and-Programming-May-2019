### Problem Statement:Define a function that take an array of integers A, and an integer K and returns the longest possible sub-set of A i.e A' such that the sum of no two elements in A' is divisible by K.First line in input contains the length of A and the integer K. Second line of input contains len(A) space-separated integers.Output must contain the length of A' list
### Testcases:
* Sample Input: 4  3
1 7 2 4

Sample Output : 3

```python
def prob3(n,k):
    a=[]
    count=0
    for i in range(0,n):
        m=int(input())
        a.append(m)
    #print(a)
    for i in range(0,n):
        for j in range(i+1,n):
            if((a[i]+a[j])%3==0):
                count+=1
    print('Longest possible subset of A :',count)
prob3(4,3)
```

## Problem 3:
### Problem statement :Define a function to add new row data to a csv file
### Constraints:
### Test cases:

```python
import pandas as pd
def addRowCSV(filename,rowdata): # adding a row means ading a list into file
    df=pd.read_csv(filename)
    df.loc[len(df)]=rowdata #adding new data 
    df.to_csv(filename,index=False) # adding data to a csv

    return df
rowdata=['0101011874','Hyderabad','63748','44237','72439','27982','61432','74833','79822','94382','90831']
addRowCSV('DataFiles/income.csv',rowdata)
```

```python
import pandas as pd
def updateRowCsv(filename,rowindex,rowdata):
    df=pd.read_csv(filename)
    df.loc[rowindex]=rowdata
    df.to_csv(filename)
    return df

rowdata=['0101011874','Banglore','63748','44237','72439','27982','61432','74833','79822','94382','90831']
updateRowCsv('DataFiles/income.csv',6,rowdata)
```
